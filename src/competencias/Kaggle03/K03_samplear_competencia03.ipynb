```python
%%time
%%sql
SELECT foto_mes, 
COUNT(*) as cantidad 
FROM competencia_03
GROUP BY foto_mes
```

```python
%%time
%%sql
SELECT numero_de_cliente, 
COUNT(*) as cantidad 
FROM competencia_03
GROUP BY numero_de_cliente
having cantidad between 5 and 15
order by cantidad desc
```

# Filtro archivos

## Archivo de 100 clientes


```python
%%time
%%sql
create or replace table competencia_03_100 as
SELECT * from competencia_03 where numero_de_cliente in 
(select numero_de_cliente
FROM competencia_03
ORDER BY RANDOM()
LIMIT 100)
```

```python
%%time
%%sql
SELECT * from competencia_03_100
```

## Archivo de 1000 Clietes

```python
%%time
%%sql
create or replace table competencia_03_1000 as
SELECT * from competencia_03 where numero_de_cliente in 
(select numero_de_cliente
FROM competencia_03
ORDER BY RANDOM()
LIMIT 1000)
```
